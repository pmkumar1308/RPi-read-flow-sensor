```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from datetime import date


```

```python
# file_name = input("Enter file name suffix (after Rpi_data_readings_cyclic_flow_rate): ")



def plot_data(df):
    df["Mass"] = df["Mass"] * 1e6
    df["Desired Mass"] = df["Desired Mass"] * 1e6

    data = df.to_numpy()

    # df["Time"] = pd.Series(list(range(len(df))))
    # df.plot(x="Time", y=header_names[1:-3], label = ["Actual Value(l/min)", "Set Value(l/min)"], ylabel = "Mass flow rate l/min", figsize=(15,5))
    # df.plot(x="Time", y="Pressure", ylabel = "Pressure kPa", figsize = (15,5))
    # df.plot(x="Time", y=["Mass", "Desired Mass"], label =["Integrated Mass mg", "Desired Mass mg"], ylabel = "Mass in mg", figsize = (15,5))

    fig, axs = plt.subplots(3,1, sharex = True, figsize=(15,15))
    st = fig.suptitle("Data plotted on", fontsize="x-large")
    axs[0].plot(data[:,0], data[:,4], label = "Integerated mass mg")
    axs[0].plot(data[:,0], data[:,5], label = "Desired Mass mg")
    axs[0].set_xlabel("Time in seconds")
    axs[0].set_ylabel("Mass in mg")
    axs[0].set_title("Integrated mass vs Desired mass")
    axs[0].legend()

    axs[1].plot(data[:,0], data[:,1], label = "Actual Value(l/min)")
    axs[1].plot(data[:,0], data[:,2], label = "Set Value(l/min)")
    axs[1].set_xlabel("Time in seconds")
    axs[1].set_ylabel("Volume flow rate in l/min")
    axs[1].set_title("Actual vs Set value of the flow rate")
    axs[1].legend()

    axs[2].plot(data[:,0], data[:,3], label = "Pressure in kPa")
    axs[2].set_xlabel("Time in seconds")
    axs[2].set_ylabel("Pressure in kPa")
    axs[2].set_title("Measured pressure")
    axs[2].legend()
    
    st.set_y(0.95)
    fig.subplots_adjust(top=0.85)

    plt.show()


    print("The desired maximum mass is:",90 ," mg")
    mass_max_error = df["Mass"].max() - 90
    print("The error in reaching the desired mass:", mass_max_error, " mg")
    return mass_max_error
```

```python
def plot_csv(file_path,del_t):
    mass_error = []
    time_delay = []
    time_delay.append(del_t)
    print("Data from time delay of", del_t," s\n\n")
    s = str(del_t)

    header_names = ['Time', 'Actual Value' , 'Set Value', 'Pressure', 'Mass', 'Desired Mass']
    df = pd.read_csv(file_path, names = header_names)
    mass_temp = plot_data(df)
```

```python
file_path_name = "./Data_from_PropValves/Rpi_data_readings_cyclic_flow_rate_001delt.csv"

plot_csv(file_path_name,0.001)
```
