```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from datetime import date


```

```python
# file_name = input("Enter file name suffix (after Rpi_data_readings_cyclic_flow_rate): ")



def plot_data(df):   
    max_range = 200
    data = df.to_numpy()
    fig= plt.figure(figsize=(15,8))
    plt.plot(data[0:max_range,0],data[0:max_range,1], label = "Actual Value(l/min)")
    plt.plot(data[0:max_range,0], data[0:max_range,2],label = "Set Value(l/min)")
    plt.xlabel("Time in seconds")
    plt.ylabel("Volume flow rate in l/min")
    plt.title("Actual vs Set value of the flow rate")
    plt.legend()
    plt.show()
```

```python
def plot_csv(file_path,del_t, target_flow, kp, kd, ki):
    mass_error = []
    time_delay = []
    time_delay.append(del_t)
    print("Data from time delay of", del_t," s\n\n")
    print("Target flow: " + str(target_flow)+" l/min")
    print("Proportional Gain: ", kp)
    print("Deivative Gain: ", kd)
    print("Integral Gain: ", ki)
    s = str(del_t)

    header_names = ['Time', 'Actual Flow Value' , 'Target Flow Value']
    df = pd.read_csv(file_path, names = header_names)
    plot_data(df)
```
