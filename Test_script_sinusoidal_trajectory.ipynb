```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
TIME_DURATION = 4
TargetMass = 80
class sinosoidalTrajectory:
    def __init__(self,Mass_initial, Mass_to_reach):
        self.initial_Mass = Mass_initial
        self.Final_mass = Mass_to_reach
        self.completion_time = TIME_DURATION
        self.amplitude = (Mass_to_reach - Mass_initial)/2
        self.frequency = 1 / self.completion_time

    def getMass(self,t):
        return self.amplitude*np.sin(2 * np.pi * self.frequency * t - np.pi/2 ) + (self.initial_Mass + self.Final_mass)/2 #/ self.completion_time
    
    def getFlow(self,t):
        return self.amplitude*np.cos(2 * np.pi * self.frequency * t - np.pi/2) 
```

```python
desiredMassWhole = []
desiredFlowWhole = []
sinTraj = sinosoidalTrajectory(20, 60)
time_step = 0.01
t=0
while t <=2*TIME_DURATION-time_step:
    desiredMass = sinTraj.getMass(t)
    desiredFlow = sinTraj.getFlow(t)
    desiredMassWhole.append(desiredMass)
    desiredFlowWhole.append(desiredFlow)
    t = t+time_step
    
```

```python
desiredMassArray = np.array(desiredMassWhole)
desiredFlowArray = np.array(desiredFlowWhole)
timeArray = np.arange(0,2*TIME_DURATION,time_step)

plt.plot(timeArray,desiredMassArray)
plt.plot(timeArray,desiredFlowArray)
plt.show()
```
